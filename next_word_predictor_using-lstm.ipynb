```python
text=[]
with open('/kaggle/input/new-data/data.txt','r',encoding='utf-8') as f:
    while(f.readline()):
        text.append(f.readline())
```

```python
text
```

```python
input_text=[]
for i in range(0,len(text),2):
    txt=''
    txt=txt + text[i].replace('\t',' ').replace('\n','')
    input_text.append(txt)
```

```python
input_text
```

```python
#Removing special characters
def full_stop(text):
  return text.replace(".","").replace("?","").lower()
```

```python
new_input=[]
for i in input_text:
    new_input.append(full_stop(i))
```

```python
new_input[9]
```

```python
import tensorflow as tf
import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(new_input)
```

```python
tokenizer.word_index
```

```python
number_of_words=len(tokenizer.word_index)-1
```

```python
number_of_words
```

```python
input_sequences = []
for sentence in input_text:
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
X=padded_input_sequences[:,:-1]
```

```python
X
```

```python
y=padded_input_sequences[:,-1]
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
from keras.utils import to_categorical
y=to_categorical(y)
```

```python
y.shape
```

```python
from keras import Sequential
from keras.layers import Embedding,Dense,SimpleRNN,LSTM
```

```python
model =Sequential()
model.add(Embedding(number_of_words,300,input_shape=(18,)))
model.add(LSTM(150))
model.add(Dense(1753,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(X,y,epochs=50)
```

```python
import time
import numpy as np
text = "The movie was"

for i in range(20):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=max_len, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print()
      print(text)
      
      time.sleep(1)
```
